```python
print(df.head())
from matplotlib import pyplot as plt
df.hist()
plt.show()
```

```python
import seaborn as sns
   # create a subplot of 3 x 3
plt.subplots(3,3,figsize=(15,15))
   # Plot a density plot for each variable
for idx, col in enumerate(df.columns):
  ax = plt.subplot(3,3,idx+1)
  ax.yaxis.set_ticklabels([])
  sns.distplot(df.loc[df.Outcome == 0][col], hist=False, axlabel= False,kde_kws={'linestyle':'-','color':'green', 'label':"No Diabetes"})
  sns.distplot(df.loc[df.Outcome == 1][col], hist=False, axlabel= False,kde_kws={'linestyle':'--','color':'red', 'label':"Diabetes"})
  ax.set_title(col)
plt.subplot(3,3,9).set_visible(False)
```
